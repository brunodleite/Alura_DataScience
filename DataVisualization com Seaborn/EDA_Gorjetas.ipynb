##Teste de hipÃ³tese

**H<sup>null</sup>**

>**HipÃ³tese nula: A distribuiÃ§Ã£o da taxa da gorjeta Ã© igual nos dois grupos!**

**H<sup>alt</sup>**

>**HipÃ³tese alternativa: A distribuiÃ§Ã£o da taxa da gorjeta nÃ£o Ã© igual nos dois grupos!**



```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
sobremesa
```

```python
sem_sobremesa = gorjetas.query("sobremesa == 'NÃ£o'").porcentagem
sem_sobremesa
```

```python
#SerÃ¡ descartado a hipÃ³tese alt, pois o p-value estÃ¡ muito alto (acima de 0.05).
#Apesar dos dois grupos pagarem gorjetas diferentes entre si(quem consome/nÃ£o consome), matematicamente a diferenÃ§a Ã© insignificante
#NÃ£o podemos dizer que a populaÃ§Ã£o Ã© diferente.  

r = ranksums(sobremesa, sem_sobremesa)
r
```

```python
#O p-value representa a probabilidade daquela amostra ter acontecido dentro da populaÃ§Ã£o. Se a chance Ã© pequena, geralmente p-value < 0,05,
#representa que um evento muito raro aconteceu, entÃ£o optamos por descartar a hipÃ³tese nula, e dizer que ela pode nÃ£o ser verdade.

print('O valor do p-value Ã© {}'.format(r.pvalue))
```

**H<sup>null</sup>**

>**HipÃ³tese nula: A distribuiÃ§Ã£o da taxa da gorjeta Ã© igual nos dois grupos!**
