##Teste de hipótese

**H<sup>null</sup>**

>**Hipótese nula: A distribuição da taxa da gorjeta é igual nos dois grupos!**

**H<sup>alt</sup>**

>**Hipótese alternativa: A distribuição da taxa da gorjeta não é igual nos dois grupos!**



```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
sobremesa
```

```python
sem_sobremesa = gorjetas.query("sobremesa == 'Não'").porcentagem
sem_sobremesa
```

```python
#Será descartado a hipótese alt, pois o p-value está muito alto (acima de 0.05).
#Apesar dos dois grupos pagarem gorjetas diferentes entre si(quem consome/não consome), matematicamente a diferença é insignificante
#Não podemos dizer que a população é diferente.  

r = ranksums(sobremesa, sem_sobremesa)
r
```

```python
#O p-value representa a probabilidade daquela amostra ter acontecido dentro da população. Se a chance é pequena, geralmente p-value < 0,05,
#representa que um evento muito raro aconteceu, então optamos por descartar a hipótese nula, e dizer que ela pode não ser verdade.

print('O valor do p-value é {}'.format(r.pvalue))
```

**H<sup>null</sup>**

>**Hipótese nula: A distribuição da taxa da gorjeta é igual nos dois grupos!**

## Teste de hipótese

>**A distribuição do valor da conta não é igual no sábado e no domingo**
